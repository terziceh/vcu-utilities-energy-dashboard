Loading of the Table into notebook in order to clean the N/As

```python
df = pd.read_excel(
    "EDA for FY19-FY25.xlsx",
    sheet_name="Missing_Buildings",
    engine="openpyxl",
    keep_default_na=True,
    na_values=["", "#N/A", "#NA", "N/A", "NA"]
)

df.head()
```

```python
missing_rows = df[df["Building_Name"].isna()]

missing_rows
```

```python
valid = df[df["Building_Name"].notna()].copy()

valid
```

```python
counts = (
    valid.groupby(
        ["Account", "Bldg", "Building_Name", "Status", "Campus"]
    )
    .size()
    .reset_index(name="count")
)

counts
```

```python
counts_sorted = counts.sort_values(
    ["Account", "Bldg", "count"],
    ascending=[True, True, False]
)

mode_map = counts_sorted.drop_duplicates(["Account", "Bldg"])

mode_map
```

```python
# Drop any leftover mode columns from earlier attempts
mode_cols = [c for c in df.columns if c.endswith("_mode")]
df = df.drop(columns=mode_cols, errors="ignore")

df.columns[df.columns.str.endswith("_mode")]
```

```python
df = df.merge(
    mode_map[["Account", "Bldg", "Building_Name", "Status", "Campus"]],
    on=["Account", "Bldg"],
    how="left",
    suffixes=("", "_mode")
)
```

```python
missing_mask = df["Building_Name"].isna()

df.loc[missing_mask, "Building_Name"] = df.loc[missing_mask, "Building_Name_mode"]
df.loc[missing_mask, "Status"] = df.loc[missing_mask, "Status_mode"]
df.loc[missing_mask, "Campus"] = df.loc[missing_mask, "Campus_mode"]

df = df.drop(columns=["Building_Name_mode", "Status_mode", "Campus_mode"])
```

```python
df[df["Building_Name"].isna()][["Account","Bldg","Status","Campus"]].head(20)
```

```python
# account-level mode map for rows that have valid Building_Name
valid2 = df[df["Building_Name"].notna()].copy()

acct_counts = (
    valid2.groupby(["Account", "Building_Name", "Status", "Campus"])
          .size()
          .reset_index(name="count")
)

acct_counts = acct_counts.sort_values(
    ["Account", "count"],
    ascending=[True, False]
)

acct_mode_map = acct_counts.drop_duplicates(["Account"])[
    ["Account", "Building_Name", "Status", "Campus"]
].rename(columns={
    "Building_Name": "Building_Name_acct",
    "Status": "Status_acct",
    "Campus": "Campus_acct"
})
```

```python
# merge account-level fallback
df = df.merge(acct_mode_map, on="Account", how="left")

still_missing = df["Building_Name"].isna()

df.loc[still_missing, "Building_Name"] = df.loc[still_missing, "Building_Name_acct"]
df.loc[still_missing, "Status"] = df.loc[still_missing, "Status_acct"]
df.loc[still_missing, "Campus"] = df.loc[still_missing, "Campus_acct"]

# drop helper columns
df = df.drop(columns=["Building_Name_acct", "Status_acct", "Campus_acct"])
```
